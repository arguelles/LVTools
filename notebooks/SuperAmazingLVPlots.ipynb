```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import rc, rcParams
import numpy as np
rc('text',usetex=True)
rc('font',**{'family':'serif','serif':['Computer Modern'], 'size' : 18})
cols = ['#29A2C6','#FF6D31','#FFCB18','#73B66B','#EF597B', '#333333']
GeV = 1.0e9
```

# NO LV

```python
NoLV=np.genfromtxt("../data/honda_propagated_noLV.dat")
```

```python
energies = NoLV[:,0]/GeV
numu_flux = NoLV[:,3]
numubar_flux = NoLV[:,4]
```

```python
plt.plot(energies,numu_flux)
plt.plot(energies,numubar_flux, lw = 2)
plt.loglog()

plt.xlim(1.0e4,1.0e6)
plt.title("Propagated Honda - No LV")
plt.xlabel("E [GeV]")
plt.ylabel("Flux")
plt.savefig("../plots/propagated_honda_nolv.png")
```

# LV: C=1.0e-27

```python
LV=np.genfromtxt("../data/honda_propagated.dat")
```

```python
energies = LV[:,0]/GeV
lv_numu_flux = LV[:,3]
lv_numubar_flux = LV[:,4]
```

```python
plt.plot(energies,lv_numu_flux)
plt.plot(energies,lv_numubar_flux)
plt.loglog()
plt.xlim(1.0e4,1.0e6)
plt.ylim(1.0e-26,1.0e-14)
plt.title("Propagated Honda - cmutau = 1.0e-27")
plt.xlabel("E [GeV]")
plt.ylabel("Muon Flux")
plt.savefig("../plots/propagated_honda_c_m27_numu.png")
```

```python
energies = LV[:,0]/GeV
lv_nutau_flux = LV[:,5]
lv_nutaubar_flux = LV[:,6]
```

```python
plt.plot(energies,lv_nutau_flux, label = "nutau")
plt.plot(energies,lv_nutaubar_flux, label = "nutaubar")
plt.loglog()
plt.xlim(1.0e4,1.0e6)
plt.ylim(1.0e-25,1.0e-15)
plt.title("Propagated Honda - cmutau = 1.0e-27")
plt.xlabel("E [GeV]")
plt.ylabel("Tau Flux")
plt.legend(loc = "upper right",fancybox=True, fontsize = 14)
plt.savefig("../plots/propagated_honda_c_m27_nutau.png")
```

# Ratio

```python
plt.plot(energies,
         (lv_numu_flux+lv_numubar_flux)/(numu_flux+numubar_flux),
        lw = 2)
plt.semilogx()
plt.ylim(0,5)
plt.xlim(1.0e4,5.0e5)
plt.xlabel("E [GeV]")
plt.ylabel("Ratio")
plt.savefig("../plots/propagated_honda_ratio.png")
```
